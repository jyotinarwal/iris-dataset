```python
import numpy as np
import pandas as pd
#load the data 
iris=pd.read_csv("Iris.csv")
iris.head()
```

```python
iris['Species'].value_counts()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#IRIS FEATURES
iris.plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm')
```

```python
#seaborn jointplot shows bivariate scatterplots and univariate histograms
import seaborn as sns
sns.jointplot(data=iris,x='SepalLengthCm',y='SepalWidthCm',size=10)
```

```python
iris.shape
```

```python
iris.info()
```

```python
#use seaborn FacetGrid to color the scatterplot by species
sns.FacetGrid(iris,hue='Species',size=5).map(plt.scatter,'SepalLengthCm','SepalWidthCm').add_legend()
```

```python
#boxplot
sns.boxplot(x='Species',y='PetalLengthCm',data=iris)
```

```python
#violin plot combines the benifits of the previous 2 plots and simplifies them 
sns.violinplot(x='Species',y='PetalLengthCm',data=iris,size=6)
```

```python
#pairplot
sns.pairplot(data=iris,hue='Species',size=3)
```

```python
sns.factorplot('SepalLengthCm',data=iris,hue="Species",kind="count")
```
